```python
from keras.datasets import imdb
import numpy as np
```

```python
#save np.load
np_load_old=np.load

#modify the default parameter of np.load
np.load=lambda *a,**k: np_load_old(*a,allow_pickle=True,**k)

#call load data with allow_pickle setting to true
(X_t,Y_t),(X_T,Y_T)=imdb.load_data(num_words=10000)

#resore np.load for future normal use
np.load=np_load_old
```

```python
### VECTORISATION
def vectorise(inp):
    out=np.zeros((len(inp),10000))
    for i,idx in enumerate(inp):
        out[i,idx]=1
    return out
    
```

```python
print(X_t)
```

```python
X_train=vectorise(X_t)
X_test=vectorise(X_T)
```

```python
X_train,X_test
```

```python
Y_train=np.array(Y_t).astype('float32')
Y_test=np.array(Y_T).astype('float32')
```

```python
Y_train
```

```python
Y_test
```

## MODEL

```python
from keras import models
from keras.layers import Dense
```

```python
model=models.Sequential()
```

```python
#model definition
model.add(Dense(16,activation='relu',input_shape=(10000,)))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
## model compile
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
X_val=X_train[:5000]
X_train_new=X_train[5000:]
Y_val=Y_train[:5000]
Y_train_new=Y_train[5000:]
```

```python
## fit
hist=model.fit(X_train_new,Y_train_new,epochs=20,batch_size=256,validation_data=(X_val,Y_val))
```

```python
hist.history
```

```python
import matplotlib.pyplot as plt
```

```python
h_dict=hist.history
plt.plot(h_dict['acc'],label="Accuracy")
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(h_dict['loss'],label="Cost function")
plt.xlabel('epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
training_loss, training_acc=model.evaluate(X_train,Y_train)
print("Training Loss: ",training_loss)
print("Training Accuracy: ",training_acc)
```

```python
testing_loss, testing_acc=model.evaluate(X_test,Y_test)
print("Testing Loss: ",training_loss)
print("Testing Accuracy: ",testing_acc)
```
